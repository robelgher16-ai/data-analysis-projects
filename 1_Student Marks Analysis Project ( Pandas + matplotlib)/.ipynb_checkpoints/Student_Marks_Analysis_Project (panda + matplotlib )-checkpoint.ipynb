## 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. Load Dataset

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

## 3. Explore Columns

```python
df.columns
```

## 4. Calculate Average Scores

```python
print(df["math score"].mean())
print(df["reading score"].mean())
print(df["writing score"].mean())
```

## 5. Visualize Average Scores

```python
df[["math score","reading score","writing score"]].mean().plot(kind="bar")
plt.title("Average Score")
plt.ylabel("value")
plt.show()
```
